## 4.4 Univariate Analysis

#### 4.4.1 GENDER COLUMN

How is distribution of Gender ?

Is gender has any impact on student's performance ?

UNIVARIATE ANALYSIS ( How is distribution of Gender ? )

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
sns.countplot(x=df['gender'],data=df,palette ='bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)
    
plt.pie(x=df['gender'].value_counts(),labels=['Male','Female'],explode=[0,0.1],autopct='%1.1f%%',shadow=True,colors=['#ff4d4d','#ff8000'])
plt.show()
```

**Insights**

* Gender has balanced data with female students are 518 (48%) and male students are 482 (52%)

## BIVARIATE ANALYSIS (Does gender have any impact in student's performance?)
  

```python

# Select only the numeric score columns
numeric_cols = ['math score', 'reading score', 'writing score']

# Group by gender and calculate the mean scores
gender_group = df.groupby('gender')[numeric_cols].mean()

# Add a new column for the total average of scores
gender_group['average'] = gender_group.mean(axis=1)


plt.figure(figsize=(10, 8))

X = ['Total Average', 'Math Average']
X_axis = np.arange(len(X))  

female_scores = [gender_group.loc['female', 'average'], gender_group.loc['female', 'math score']]
male_scores = [gender_group.loc['male', 'average'], gender_group.loc['male', 'math score']]

plt.bar(X_axis - 0.2, male_scores, 0.4, label='Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label='Female')

plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total Average vs Math Average Marks of Both Genders", fontweight='bold')
plt.legend()
plt.show()

                                             
```

**Insight**
* On average females have a better overall score than men.
* whereas males have scored higher in Maths.

### BIVARIATE ANALYSIS ( Is lunch type intake has any impact on student's performance ? )

```python
# List of score columns
score_columns = ['math score', 'reading score', 'writing score']

# Set the style
sns.set(style="whitegrid")

# Set up the figure
plt.figure(figsize=(15, 5))

# Loop through each score column
for i, col in enumerate(score_columns):
    plt.subplot(1, 3, i+1)
    sns.barplot(x='lunch', y=col, data=df)
    plt.title(f'{col.title()} by Lunch Type')
    plt.xlabel('Lunch Type')
    plt.ylabel('Average Score')

# Adjust layout
plt.tight_layout()
plt.show()
```

**Insights**
  
* Students who get Standard Lunch tend to perform better than students who got free/reduced lunch

###  BIVARIATE ANALYSIS ( Is Test prepration course has any impact on student's performance ? )

```python
plt.rcParams['figure.figsize'] = (15, 9)
plt.style.use('fivethirtyeight')
sns.countplot(df['test preparation course'], palette = 'Blues')
plt.title('Comparison of Test Preparation Course', fontweight = 30, fontsize = 20)
plt.xlabel('Degree')
plt.ylabel('count')
plt.show()
```

**Insights**

* Most students dont take test preparation course

### BIVARIATE ANALYSIS ( Is Test prepration course has any impact on student's performance ? )

```python
# Set up the figure size and style
plt.figure(figsize=(15, 5))
sns.set(style="whitegrid")

# List of subjects
score_columns = ['math score', 'reading score', 'writing score']

# Create subplots using a loop
for i, col in enumerate(score_columns):
    plt.subplot(1, 3, i+1)
    sns.barplot(x='test preparation course', y=col, data=df, ci='sd', palette='Set2')
    plt.title(f'{col.title()} by Test Preparation Course')
    plt.xlabel('Test Preparation Course')
    plt.ylabel('Average Score')

plt.tight_layout()
plt.show();
```

**Insights**
  
* Students who have completed the Test Prepration Course have scores higher in all three categories than those who haven't taken the course.
